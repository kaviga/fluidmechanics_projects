```python
#Dibujando las líneas de corriente por tiempos entre 1s y 25s del campo de velocidades bidimensional no-estacionario, en el punto (1,1).
import math
import matplotlib as mt
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

x=np.linspace(0,2,50)
entries=[]
plt.figure(figsize=(10, 10))
plt.xlim(0, 2)
plt.ylim(-1, 2)

for n in range (1,26):
    y =  1/(x ** (1/2*n+1))
    plt.plot(x,y,'r')
    entries.append('time='+str(n)+'s')


plt.grid()
plt.legend(entries)
plt.show()
```
